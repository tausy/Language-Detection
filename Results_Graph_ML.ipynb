## Plotting the Results for 34 European Languages

```python
import matplotlib as plt
%matplotlib inline 
import pandas as pd
import numpy as np
result = pd.read_csv('/Users/ashwinkumar/Desktop/EU_languages.csv')
ax = result.groupby("English")['Percentage_Accuracy'].mean().plot(kind='bar',title ="Match Score for 34 Languages", figsize=(12,7),legend=True, fontsize=12, width = 0.75, color = 'black', linewidth = '5' )
ax.set_xlabel("Languages",fontsize=14)
ax.set_ylabel("% Correct Guesses",fontsize=14)
ax.set_yticks(np.arange(0, 110, step=10))
ax.figure.savefig('/Users/ashwinkumar/Desktop/Result_All_Languages.png',bbox_inches='tight')

```
